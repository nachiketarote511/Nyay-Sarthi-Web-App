```python
import pandas as pd
import json
```

```python
df = pd.read_parquet(
    "../../data/processed/classification_dataset_final.parquet"
)

print(df.shape)

df.head()
```

```python
df = df[
    [
        "legal_summary",
        "category"
    ]
].copy()
```

```python
llm_dataset = []

for _, row in df.iterrows():

    llm_dataset.append(
        {
            "instruction":
                "Classify the legal case into the correct legal category.",

            "input":
                row["legal_summary"],

            "output":
                row["category"]
        }
    )
```

```python
print(
    len(llm_dataset)
)

llm_dataset[0]
```

```python
with open(
    "../../data/processed/legal_llm_dataset.json",
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        llm_dataset,
        f,
        ensure_ascii=False,
        indent=2
    )
```

```python
import os

size_mb = (
    os.path.getsize(
        "../../data/processed/legal_llm_dataset.json"
    )
    /
    1024
    /
    1024
)

print(
    round(size_mb,2),
    "MB"
)
```
