```python
import pandas as pd
import json

print("Libraries Loaded")
```

```python
df = pd.read_parquet(
    "../../data/processed/classification_dataset_final.parquet"
)

print(df.shape)

df.head()
```

```python
print(
    df["category"].value_counts()
)
```

```python
chat_dataset = []

for _, row in df.iterrows():

    user_prompt = f"""
Analyze the following legal case.

Case Summary:
{row['legal_summary']}
"""

    assistant_response = f"""
Legal Category: {row['category']}

This case primarily belongs to the {row['category']} domain of law.

Based on the facts and legal issues involved, the dispute is best classified under {row['category']} law.
"""

    chat_dataset.append(
        {
            "messages": [
                {
                    "role": "user",
                    "content": user_prompt.strip()
                },
                {
                    "role": "assistant",
                    "content": assistant_response.strip()
                }
            ]
        }
    )

print(len(chat_dataset))

print(chat_dataset[0])
```

```python
output_path = (
    "../../data/processed/qwen_chat_dataset.json"
)

with open(
    output_path,
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        chat_dataset,
        f,
        ensure_ascii=False,
        indent=2
    )

print(
    f"Saved: {output_path}"
)
```
