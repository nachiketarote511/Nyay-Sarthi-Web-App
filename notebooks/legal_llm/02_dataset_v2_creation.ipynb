```python
import pandas as pd
import json
import random
```

```python
df = pd.read_parquet(
    "../../data/processed/classification_dataset_final.parquet"
)

print(df.shape)
```

```python
query_templates = [

    "Analyze this legal case.",

    "Identify the legal category.",

    "What type of legal dispute is this?",

    "Provide legal analysis of the following case.",

    "Determine the legal domain of this case.",

    "Classify this legal matter.",

    "Explain the legal nature of this dispute.",

    "Which area of law does this case belong to?"

]
```

```python
def build_output(category):

    return f"""
Legal Category: {category}

This case primarily belongs to the {category} domain of Indian law.

The facts and legal issues discussed in the summary are most closely related to {category} matters.
"""
```

```python
llm_dataset_v2 = []

for _, row in df.iterrows():

    instruction = random.choice(
        query_templates
    )

    llm_dataset_v2.append(

        {
            "instruction":
                instruction,

            "input":
                row["legal_summary"],

            "output":
                build_output(
                    row["category"]
                )
        }

    )
```

```python
print(
    len(llm_dataset_v2)
)

llm_dataset_v2[0]
```

```python
with open(

    "../../data/processed/legal_llm_dataset_v2.json",

    "w",

    encoding="utf-8"

) as f:

    json.dump(

        llm_dataset_v2,

        f,

        ensure_ascii=False,

        indent=2

    )
```
