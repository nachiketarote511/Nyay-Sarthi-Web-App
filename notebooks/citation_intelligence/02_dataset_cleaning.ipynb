```python
import pandas as pd

df = pd.read_parquet(
    "../../data/processed/citation dataset/citation_dataset.parquet"
)

print(df.shape)
```

```python
df["cited_case"] = (
    df["cited_case"]
    .astype(str)
    .str.strip()
)
```

```python
df = df[
    df["cited_case"]
    .str.contains(
        "vs",
        case=False,
        na=False
    )
]
```

```python
df = df[
    df["cited_case"]
    .str.len() > 20
]
```

```python
bad_words = [

    "held",
    "section",
    "article",
    "constitution",
    "power conferred",
    "reasonable relation",
    "therefore",
    "provided that"

]

for word in bad_words:

    df = df[
        ~df["cited_case"]
        .str.lower()
        .str.contains(
            word,
            na=False
        )
    ]
```

```python
df = df.drop_duplicates(
    subset=[
        "citing_case",
        "cited_case"
    ]
)

print(df.shape)
```

```python
print(
    df["cited_case"]
    .value_counts()
    .head(20)
)
```

```python
df.to_parquet(
    "../../data/processed/citation dataset/citation_dataset_cleaned.parquet",
    index=False
)

print("Saved")
```
