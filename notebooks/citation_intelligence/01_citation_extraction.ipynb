```python
import pandas as pd
import re
from tqdm import tqdm

tqdm.pandas()
```

```python
df = pd.read_parquet(
    "../../data/processed/clean_cases.parquet"
)

print(df.shape)

df.head()
```

```python
df.columns
```

```python
def normalize_case_name(name):

    name = str(name)

    name = name.replace("_", " ")

    name = re.sub(
        r"_\d+$",
        "",
        name
    )

    return name.lower()
```

```python
case_names = (
    df["case_name"]
    .apply(normalize_case_name)
    .tolist()
)

print(len(case_names))
```

```python
sample_df = df.head(500).copy()

print(sample_df.shape)
```

```python
sample_df["normalized_case"] = (
    sample_df["case_name"]
    .apply(normalize_case_name)
)
```

```python
citations = []

for _, row in tqdm(
    sample_df.iterrows(),
    total=len(sample_df)
):

    source_case = row["case_name"]

    text = str(
        row["legal_text"]
    ).lower()

    for candidate in case_names:

        if candidate == row["normalized_case"]:
            continue

        if len(candidate) < 25:
            continue

        if candidate in text:

            citations.append(
                (
                    source_case,
                    candidate
                )
            )
```

```python
citation_df = pd.DataFrame(
    citations,
    columns=[
        "source_case",
        "cited_case"
    ]
)

print(citation_df.shape)

citation_df.head()
```
