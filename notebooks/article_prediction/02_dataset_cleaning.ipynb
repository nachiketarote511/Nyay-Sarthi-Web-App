```python
import pandas as pd

df = pd.read_parquet(
    "../../data/processed/article_prediction_dataset.parquet"
)

print(df.shape)

counts = df["article"].value_counts()

valid_articles = counts[
    counts >= 20
].index

df = df[
    df["article"].isin(valid_articles)
]

print(df.shape)

print(
    df["article"].value_counts()
)
```

```python
df.to_parquet(
    "../../data/processed/article_prediction_dataset_cleaned.parquet",
    index=False
)

print("Saved")
```
