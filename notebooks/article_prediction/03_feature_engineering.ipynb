```python
import pandas as pd

df = pd.read_parquet(
    "../../data/processed/artical prediction/article_prediction_dataset_cleaned.parquet"
)

print(df.shape)

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["article_encoded"] = le.fit_transform(
    df["article"]
)

print(
    len(le.classes_)
)

print(
    le.classes_
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(

    max_features=10000,

    stop_words="english",

    ngram_range=(1,2)

)
```

```python
X = tfidf.fit_transform(
    df["legal_summary"]
)

y = df["article_encoded"]

print(X.shape)
print(y.shape)
```

```python
import joblib

joblib.dump(
    tfidf,
    "../../models/article_predictor/tfidf_vectorizer.pkl"
)

joblib.dump(
    le,
    "../../models/article_predictor/label_encoder.pkl"
)

print("Saved")
```

```python
import scipy.sparse

scipy.sparse.save_npz(
    "../../data/processed/article_features.npz",
    X
)

joblib.dump(
    y,
    "../../models/article_predictor/article_labels.pkl"
)

print("Features Saved")
```
