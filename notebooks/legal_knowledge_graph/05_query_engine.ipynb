```python
def get_related_cases(case_name):

    if case_name not in G:
        return []

    neighbors = list(
        G.neighbors(case_name)
    )

    related = set()

    for n in neighbors:

        for second in G.neighbors(n):

            if second != case_name:
                related.add(second)

    return list(related)[:20]
```
