```python
import pandas as pd
import networkx as nx
```

```python
edges_df = pd.read_parquet(
    "../../data/processed/graph/kg_edges.parquet"
)

print(edges_df.shape)
```

```python
G = nx.Graph()

print("Graph Created")
```

```python
for _, row in edges_df.iterrows():

    G.add_edge(
        row["source"],
        row["target"],
        relation=row["relation"]
    )

print("Edges Added")
```

```python
print("Nodes :", G.number_of_nodes())

print("Edges :", G.number_of_edges())
```

```python
import joblib

joblib.dump(
    G,
    "../../models/legal_knowledge_graph/knowledge_graph.pkl"
)

print("Graph Saved")
```

```python
print(
    "Connected Components:",
    nx.number_connected_components(G)
)

print(
    "Density:",
    nx.density(G)
)
```

04_graph_analytics.ipynb


```python
degrees = dict(G.degree())

top_nodes = sorted(
    degrees.items(),
    key=lambda x: x[1],
    reverse=True
)[:20]

for node, degree in top_nodes:
    print(node, degree)
```

```python
centrality = nx.degree_centrality(G)

top_central = sorted(
    centrality.items(),
    key=lambda x: x[1],
    reverse=True
)[:20]

for node, score in top_central:
    print(node, score)
```

```python
category_nodes = [
    'Constitutional',
    'Civil',
    'Criminal',
    'Taxation',
    'Labour',
    'Property',
    'Administrative',
    'Corporate'
]

for node in category_nodes:

    if node in G:

        print(
            node,
            G.degree(node)
        )
```

```python
import pandas as pd

analytics_df = pd.DataFrame(
    top_nodes,
    columns=[
        "node",
        "degree"
    ]
)

analytics_df.to_csv(
    "../../reports/kg_top_nodes.csv",
    index=False
)
```

05_query_engine.ipynb


```python
def search_case(keyword):

    results = []

    for node in G.nodes():

        if keyword.lower() in str(node).lower():

            results.append(node)

    return results[:20]
```

```python
print("Nodes :", G.number_of_nodes())
print("Edges :", G.number_of_edges())
```

```python
matches = [
    node
    for node in G.nodes()
    if "Gopalan" in str(node)
]

print("Matches:", len(matches))

for m in matches[:20]:
    print(m)
```

```python
for node in G.nodes():

    if "A_K_Gopalan" in str(node):

        print(node)
```

```python
search_case("Gopalan")
```

```python
case = "A_K_Gopalan_vs_The_State_Of_Madras_Union_Of_India__on_19_May_1950_1"

neighbors = list(
    G.neighbors(case)
)

print(neighbors)
```
