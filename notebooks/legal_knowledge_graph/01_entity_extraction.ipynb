```python
import pandas as pd

print("Libraries Loaded")
```

```python
df = pd.read_parquet(
    "../../data/processed/classification dataset/classification_dataset_final.parquet"
)

print(df.shape)

df.head()
```

```python
print(df.columns.tolist())
```

```python
case_entities = pd.DataFrame()

case_entities["case_name"] = df["case_name"]

print(case_entities.shape)
```

```python
category_entities = pd.DataFrame()

category_entities["category"] = df["category"]

category_entities = category_entities.drop_duplicates()

print(category_entities.shape)
```

```python
year_entities = pd.DataFrame()

year_entities["year"] = df["year"]

year_entities = year_entities.drop_duplicates()

print(year_entities.shape)
```

```python
article_df = pd.read_parquet(
    "../../data/processed/artical prediction/article_prediction_dataset_cleaned.parquet"
)

print(article_df.shape)
```

```python
article_entities = pd.DataFrame()

article_entities["article"] = article_df["article"]

article_entities = article_entities.drop_duplicates()

print(article_entities.shape)
```

```python
case_entities.to_parquet(
    "../../data/processed/graph/kg_case_entities.parquet",
    index=False
)

category_entities.to_parquet(
    "../../data/processed/graph/kg_category_entities.parquet",
    index=False
)

year_entities.to_parquet(
    "../../data/processed/graph/kg_year_entities.parquet",
    index=False
)

article_entities.to_parquet(
    "../../data/processed/graph/kg_article_entities.parquet",
    index=False
)

print("All Entity Files Saved")
```

```python
print("Cases     :", len(case_entities))
print("Categories:", len(category_entities))
print("Years     :", len(year_entities))
print("Articles  :", len(article_entities))
```
