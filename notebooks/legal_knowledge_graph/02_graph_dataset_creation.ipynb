```python
import pandas as pd
```

```python
df = pd.read_parquet(
    "../../data/processed/classification dataset/classification_dataset_final.parquet"
)

article_df = pd.read_parquet(
    "../../data/processed/artical prediction/article_prediction_dataset_cleaned.parquet"
)

print(df.shape)
print(article_df.shape)
```

```python
case_category_edges = df[
    ["case_name", "category"]
].copy()

case_category_edges.columns = [
    "source",
    "target"
]

case_category_edges["relation"] = "BELONGS_TO"

print(case_category_edges.shape)
```

```python
case_year_edges = df[
    ["case_name", "year"]
].copy()

case_year_edges.columns = [
    "source",
    "target"
]

case_year_edges["relation"] = "FILED_IN"

case_year_edges["target"] = (
    case_year_edges["target"]
    .astype(str)
)

print(case_year_edges.shape)
```

```python
case_article = pd.merge(
    df[
        [
            "case_name",
            "legal_summary"
        ]
    ],
    article_df,
    on="legal_summary",
    how="inner"
)

print(case_article.shape)
```

```python
case_article_edges = case_article[
    [
        "case_name",
        "article"
    ]
].copy()

case_article_edges.columns = [
    "source",
    "target"
]

case_article_edges["relation"] = "INVOLVES"

print(case_article_edges.shape)
```

```python
edges_df = pd.concat(
    [
        case_category_edges,
        case_year_edges,
        case_article_edges
    ],
    ignore_index=True
)

print(edges_df.shape)

edges_df.head()
```

```python
edges_df.to_parquet(
    "../../data/processed/graph/kg_edges.parquet",
    index=False
)

print("Knowledge Graph Edges Saved")
```
