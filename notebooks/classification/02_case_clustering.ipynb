```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
```

```python
df = pd.read_parquet(
    "../../data/processed/summaries_only.parquet"
)

embeddings = np.load(
    "../../data/embeddings/embeddings.npy"
)

print(df.shape)
print(embeddings.shape)
```

```python
N_CLUSTERS = 15
```

```python
kmeans = KMeans(
    n_clusters=15,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(
    embeddings
)

df["cluster"] = clusters
```

```python
print(
    df["cluster"]
    .value_counts()
    .sort_index()
)
```

```python
df.to_parquet(
    "../../data/processed/cases_with_clusters.parquet",
    index=False
)

print("Saved")
```

```python
for cluster_id in sorted(
    df["cluster"].unique()
):

    print("\n")
    print("="*100)

    print(
        "CLUSTER:",
        cluster_id
    )

    print("="*100)

    sample = df[
        df["cluster"] == cluster_id
    ].head(5)

    for _, row in sample.iterrows():

        print()

        print(
            row["case_name"]
        )
```
