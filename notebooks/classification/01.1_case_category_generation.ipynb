```python
import pandas as pd
import numpy as np

df = pd.read_parquet(
    "../data/processed/summaries_only.parquet"
)

print(df.shape)
```

```python
def assign_category(text):

    text = str(text).lower()

    if any(word in text for word in [
        "article 14",
        "article 19",
        "article 21",
        "fundamental right",
        "constitution",
        "constitutional"
    ]):
        return "Constitutional"

    elif any(word in text for word in [
        "murder",
        "conviction",
        "criminal",
        "ipc",
        "offence",
        "accused",
        "bail"
    ]):
        return "Criminal"

    elif any(word in text for word in [
        "property",
        "land",
        "tenant",
        "landlord",
        "mortgage",
        "ownership"
    ]):
        return "Property"

    elif any(word in text for word in [
        "contract",
        "agreement",
        "specific performance",
        "broker",
        "agency",
        "commission"
    ]):
        return "Contract"

    elif any(word in text for word in [
        "tax",
        "income tax",
        "assessment",
        "revenue",
        "deduction"
    ]):
        return "Tax"

    elif any(word in text for word in [
        "service",
        "promotion",
        "government servant",
        "disciplinary",
        "department"
    ]):
        return "Service"

    elif any(word in text for word in [
        "industrial dispute",
        "workman",
        "labour",
        "employee",
        "trade union"
    ]):
        return "Labour"

    elif any(word in text for word in [
        "consumer",
        "deficiency",
        "consumer protection"
    ]):
        return "Consumer"

    else:
        return "Other"
```

```python
df["category"] = df[
    "legal_summary"
].apply(assign_category)

df["category"].value_counts()
```

```python
print(
    df["category"]
      .value_counts()
)
```

```python
df.to_parquet(
    "../data/processed/case_classification_dataset.parquet",
    index=False
)

print("Saved")
```
