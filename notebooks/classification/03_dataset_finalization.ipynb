```python
import pandas as pd
import numpy as np

df = pd.read_parquet(
    "../../data/processed/summaries_only.parquet"
)

print(df.shape)
df.head()
```

```python
def assign_category(text):

    text = str(text).lower()

    # Constitutional

    if any(word in text for word in [
        "article 14",
        "article 19",
        "article 21",
        "fundamental right",
        "constitution",
        "constitutional",
        "freedom of speech",
        "preventive detention"
    ]):
        return "Constitutional"

    # Criminal

    elif any(word in text for word in [
        "murder",
        "conviction",
        "criminal",
        "offence",
        "offense",
        "accused",
        "bail",
        "sentence",
        "imprisonment",
        "the state"
    ]):
        return "Criminal"

    # Property

    elif any(word in text for word in [
        "property",
        "land",
        "tenant",
        "landlord",
        "mortgage",
        "ownership",
        "possession",
        "agricultural land"
    ]):
        return "Property"

    # Contract

    elif any(word in text for word in [
        "contract",
        "agreement",
        "specific performance",
        "broker",
        "agency",
        "commission",
        "sale deed",
        "arbitration"
    ]):
        return "Contract"

    # Family

    elif any(word in text for word in [
        "marriage",
        "wife",
        "husband",
        "widow",
        "dower",
        "inheritance",
        "will",
        "succession",
        "heir"
    ]):
        return "Family"

    # Service

    elif any(word in text for word in [
        "service",
        "promotion",
        "government servant",
        "departmental",
        "disciplinary",
        "suspension"
    ]):
        return "Service"

    # Labour

    elif any(word in text for word in [
        "industrial dispute",
        "workman",
        "worker",
        "employee",
        "labour",
        "trade union"
    ]):
        return "Labour"

    # Tax

    elif any(word in text for word in [
        "income tax",
        "tax",
        "assessment",
        "revenue",
        "deduction",
        "commissioner of income tax"
    ]):
        return "Tax"

    # Commercial

    elif any(word in text for word in [
        "company",
        "bank",
        "corporation",
        "insurance",
        "shareholder",
        "commercial"
    ]):
        return "Company_Commercial"

    else:
        return "Civil"
```

```python
df["category"] = df[
    "legal_summary"
].apply(assign_category)

df["category"].value_counts()
```

```python
distribution = (
    df["category"]
      .value_counts()
)

print(distribution)
```

```python
df.to_parquet(
    "../../data/processed/classification_dataset.parquet",
    index=False
)

print("Saved")
```
