```python
import pandas as pd

df = pd.read_parquet(
    "../data/processed/summaries_only.parquet"
)

print(df.shape)
```

```python
import torch

print(
    "GPU:",
    torch.cuda.get_device_name(0)
)

print(
    "CUDA Available:",
    torch.cuda.is_available()
)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "BAAI/bge-small-en-v1.5",
    device="cuda"
)

print("Model Loaded")
```

```python
texts = (
    df["legal_summary"]
    .fillna("")
    .tolist()
)

print(
    "Total Texts:",
    len(texts)
)
```

```python
import time

start = time.time()

embeddings = model.encode(
    texts,
    batch_size=256,
    show_progress_bar=True,
    convert_to_numpy=True,
    normalize_embeddings=True
)

end = time.time()

print(
    "Shape:",
    embeddings.shape
)

print(
    "Time:",
    round(end-start,2),
    "seconds"
)
```

```python
import numpy as np

np.save(
    "../data/processed/embeddings.npy",
    embeddings
)

print("Saved")
```

```python
import os

size_mb = (
    os.path.getsize(
        "../data/processed/embeddings.npy"
    )
    / (1024*1024)
)

print(
    round(size_mb,2),
    "MB"
)
```
