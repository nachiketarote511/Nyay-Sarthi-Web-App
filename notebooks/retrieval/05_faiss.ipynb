```python
import faiss
import numpy as np
```

```python
embeddings = np.load(
    "../data/processed/embeddings.npy"
)

print(embeddings.shape)
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatIP(
    dimension
)

index.add(
    embeddings.astype("float32")
)

print(
    "Vectors Indexed:",
    index.ntotal
)
```

```python
faiss.write_index(
    index,
    "../data/processed/faiss_index.bin"
)
```

```python
import os

size_mb = (
    os.path.getsize(
        "../data/processed/faiss_index.bin"
    ) / 1024 / 1024
)

print(size_mb)
```
